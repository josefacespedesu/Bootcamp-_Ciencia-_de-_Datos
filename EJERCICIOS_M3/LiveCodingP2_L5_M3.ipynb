<h3>2. Convertir columnas de string a datetime.</h3>

```python
df_sin_fechas["FECHA_NACIMIENTO"].value_counts()
```

```python
df_sin_fechas["FECHA_NACIMIENTO"].dtype
```

```python
s = df_sin_fechas["FECHA_NACIMIENTO"].astype("string").str.strip()
s = s.str.replace(".", "/", regex=False).str.replace("-", "/", regex=False)

df_sin_fechas["FECHA_NACIMIENTO"] = pd.to_datetime(
    s,
    errors="coerce"
)

df_sin_fechas["FECHA_NACIMIENTO"].dtype
```

```python
df_sin_fechas.dtypes
```

```python
s = df_sin_fechas["FECHA_INGRESO"].astype("string").str.strip()
s = s.str.replace(".", "/", regex=False).str.replace("-", "/", regex=False)

df_sin_fechas["FECHA_INGRESO"] = pd.to_datetime(
    s,
    errors="coerce"
)

df_sin_fechas["FECHA_INGRESO"].dtype
```

```python
df_sin_fechas.dtypes
```

<p>Esta documentación detalla los pasos realizados en el notebook para procesar el conjunto de datos de Grupos Relacionados por el Diagnóstico (GRD), enfocándose en la reducción de dimensionalidad, normalización de tipos y gestión de registros.</p>
<ol>
        <li>
            <strong>Carga Inicial de Datos</strong>

                <li><strong>Estado Inicial:</strong> El dataset original cuenta con 129 columnas.</li>
            </ul>
        </li>
    </ol>
