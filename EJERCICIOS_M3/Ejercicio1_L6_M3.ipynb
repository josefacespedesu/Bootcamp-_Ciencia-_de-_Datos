```python
import pandas as pd

# 0. Crear el dataset de productos
data = {
    'Producto': ['Laptop', 'Mouse', 'Monitor', 'Teclado', 'Laptop', 'Monitor', 'Smartphone', 'Smartphone'],
    'Tipo': ['Electrónica', 'Accesorios', 'Electrónica', 'Accesorios', 'Electrónica', 'Electrónica', 'Móvil', 'Móvil'],
    'Marca': ['Dell', 'Logitech', 'Samsung', 'Logitech', 'HP', 'LG', 'Apple', 'Xiaomi'],
    'Precio': [1200, 25, 300, 45, 1100, 280, 1000, 600]
}

df = pd.DataFrame(data)
df
```

```python
#1. Agrupar un dataset de productos por “tipo”
df_grouped = df.groupby('Tipo')
df_grouped
```

```python
#2. Mostrar cantidad de productos y precio promedio.
summary = df_grouped.agg({'Producto': 'count', 'Precio': 'mean'}).rename(columns={'Producto': 'Cantidad de Productos', 'Precio': 'Precio Promedio'})
summary
```

```python
#3. Aplicar .agg() con dos funciones combinadas.
detailed_summary = df_grouped.agg({
    'Precio': ['mean', 'max', 'min'],
    'Producto': 'count'
}).rename(columns={'Producto': 'Cantidad de Productos'})

detailed_summary
```

```python
#ASI LO HIZO LA PROFE 
# 1, 2 y 3. Agrupar por "Tipo", mostrar cantidad y precio promedio usando .agg()
# 'count' nos da la cantidad de productos y 'mean' el precio promedio
df_resumen = df.groupby('Tipo').agg({
    'Precio': ['count', 'mean']
})

df_resumen
```

```python
#4. Usar .pivot_table() para ver totales por marca y tipo.
tabla_cantidades = pd.pivot_table(
    df,
    index='Marca',
    columns='Tipo',
    aggfunc='size',
    fill_value=0
)

tabla_cantidades
```
