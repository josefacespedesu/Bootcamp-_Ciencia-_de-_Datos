```python
import pandas as pd
```

```python
df = pd.read_csv('usuarios_sucios.csv')
df.shape
df.head(10)
```

```python
#Eliminar columanas innecesarias
df_new= df.drop(columns=['columna_innecesaria1','columna_innecesaria2'])
df_new.head()
```

```python
#Asegura que todo sea texto (aunque haya números, NaN, mezclas Elimina espacios invisibles
s = df_new["fecha_registro"].astype("string").str.strip()
#deja todas la fechas iguales
s = s.str.replace(".", "/", regex=False).str.replace("-", "/", regex=False)
# cambia a formato fecha los datos ya limpiados
df_new["fecha_registro"] = pd.to_datetime(
    s,
    errors="coerce"
)

df_new.head()
```
