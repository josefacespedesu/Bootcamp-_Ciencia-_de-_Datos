```python
import pandas as pd

# 0. Crear el set de datos para los departamentos
data = {
    'Departamento': ['Ventas', 'Ventas', 'TI', 'TI', 'RRHH', 'RRHH', 'Ventas', 'TI'],
    'Empleado': ['Ana', 'Luis', 'Carlos', 'Sofía', 'Elena', 'Pablo', 'Jorge', 'Marta'],
    'Puntuación': [85, 78, 92, 95, 60, 65, 88, 91]
}

df = pd.DataFrame(data)
df
```

```python
#1. Agrupar rendimiento por departamento.
grouped_departamento = df.groupby('Departamento')
print (grouped_departamento)
```

```python
#2. Calcular promedio, mínimo y máximo de puntuación.
estadistica = grouped_departamento['Puntuación'].agg(['mean', 'min', 'max'])
estadistica
```

```python
#3. Filtrar resultados mayores a un umbral.
umbral=85
mayores_a_85=df.groupby('Departamento').filter(lambda x: x['Puntuación'].mean() > umbral)
mayores_a_85
```

```python
#4. Ordenar de mayor a menor.
ordenado = estadistica.sort_values(by='mean', ascending=False)
ordenado
```

```python
#5. Preparar reporte final listo para exportar a Excel.
reporte_final= ordenado.rename(columns=
{'mean':'Promedio Puntuación',
 'min':'Mínimo Puntuación',
 'max':'Máximo Puntuación'
 })
reporte_final
```
