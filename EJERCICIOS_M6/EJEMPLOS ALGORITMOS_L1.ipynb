```python
#Gráﬁco de dispersión con frontera de decisión entre dos clases

from sklearn.datasets import make_classiﬁcation
from sklearn.linear_model import LogisticRegression
17
import matplotlib.pyplot as plt
import numpy as np
# Datos simulados
X, y = make_classiﬁcation(n_samples=100, n_features=2, n_redundant=0,
n_classes=2, random_state=42)
model = LogisticRegression()
model.ﬁt(X, y)
# Frontera de decisión
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max,
100))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k')
plt.title("Clasiﬁcación Binaria: Spam / No spam")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

```python
#Gráﬁco de dispersión 2D con 3 clases codiﬁcadas por color:

from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=150, centers=3, random_state=0, cluster_std=1.2)
model = LogisticRegression(solver='lbfgs')
model.fit(X, y)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='Set1', edgecolor='k')
plt.title("Clasiﬁcación Multiclase: Vehículos")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

```python
#Mapa binario de etiquetas para diferentes instancias (tipo heatmap).

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

multi_label_data = pd.DataFrame({
"Imagen 1": [1, 0, 1, 0],
"Imagen 2": [0, 1, 1, 0],
"Imagen 3": [1, 1, 0, 1],
"Imagen 4": [1, 0, 1, 1]
}, index=["Persona", "Perro", "Bicicleta", "Árbol"])
sns.heatmap(multi_label_data, annot=True, cmap="YlGnBu", cbar=False)
plt.title("Clasiﬁcación Multi-etiqueta: Etiquetas por Imagen")
plt.xlabel("Instancias")
plt.ylabel("Etiquetas")
plt.show()
```

```python
#Visualización conceptual: Red neuronal para clasiﬁcación

import networkx as nx
22
import matplotlib.pyplot as plt
G = nx.DiGraph()
layers = [3, 4, 3] # entradas, ocultas, salidas
positions = {}
node = 0
# Crear capas
for i, layer_size in enumerate(layers):
    for j in range(layer_size):
        G.add_node(node)
        positions[node] = (i, -j)
        node += 1
# Conexiones
for i in range(layers[0]):
    for j in range(layers[0], layers[0] + layers[1]):
        G.add_edge(i, j)
for i in range(layers[0], layers[0] + layers[1]):
    for j in range(layers[0] + layers[1], sum(layers)):
        G.add_edge(i, j)
        nx.draw(G, pos=positions, with_labels=False, node_size=600,
        node_color='lightblue', arrows=False)
23
plt.title("Red Neuronal para Clasiﬁcación Multiclase")
plt.axis('off')
plt.show()
```
