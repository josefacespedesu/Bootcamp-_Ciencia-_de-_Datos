## **Uniforme Discreta**

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de 1000 lanzamientos de un dado justo
np.random.seed(42)
muestras = np.random.randint(1, 7, size=1000) # 1000000

# Graficamos frecuencias relativas
valores, conteos = np.unique(muestras, return_counts=True)
frecuencias = conteos / len(muestras)

print("Probabilidades estimadas:")
for v, f in zip(valores, frecuencias):
    print(f"Valor {v}: {f:.3f}")


plt.bar(valores, frecuencias)
plt.xlabel("Valor obtenido")
plt.ylabel("Frecuencia relativa")
plt.title("Distribución Uniforme Discreta (Dado justo)")
plt.xticks(valores)
plt.show()
```

* Como todos los valores son igualmente probables, la gráfica debería mostrar barras de altura similar. Si aumentamos el número de simulaciones, se verá cada vez más uniforme.

## Distribución de Bernoulli

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la Bernoulli
p = 0.7

# Simulamos 1000 experimentos
np.random.seed(42)
muestras = np.random.binomial(n=1, p=p, size=1000)

# Calculamos frecuencias
valores, conteos = np.unique(muestras, return_counts=True)
frecuencias = conteos / len(muestras)

print("Media estimada:", np.mean(muestras))
print("Varianza estimada:", np.var(muestras))


# Graficamos
plt.bar(valores, frecuencias)
plt.xticks([0, 1])
plt.xlabel("Resultado (0 = fracaso, 1 = éxito)")
plt.ylabel("Frecuencia relativa")
plt.title("Distribución Bernoulli (p = 0.7)")
plt.show()
```

Aquí estamos simulando un experimento con solo dos posibles resultados. Al repetirlo muchas veces, observamos que la proporción de éxitos se aproxima a p.

## Distribución binomial

Ejemplo: lanzar una moneda 10 veces con probabilidad de cara
p =0.7

La variable ahora es: “cuántas caras obtuve en 10 lanzamientos”.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la Binomial
n = 10
p = 0.7

# Simulamos 1000 experimentos (cada experimento = 10 ensayos)
np.random.seed(42)
muestras = np.random.binomial(n=n, p=p, size=1000)

# Graficamos histograma (frecuencia relativa)
plt.hist(muestras, bins=np.arange(-0.5, n + 1.5, 1), density=True)
plt.xlabel("Número de éxitos en 10 ensayos")
plt.ylabel("Frecuencia relativa")
plt.title("Distribución Binomial (n=10, p=0.7)")
plt.xticks(range(0, n + 1))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.7
np.random.seed(42)

# Bernoulli: n=1
bernoulli = np.random.binomial(n=1, p=p, size=1000)

# Binomial: n=10
binomial = np.random.binomial(n=10, p=p, size=1000)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Bernoulli (solo 0 o 1)
valores, conteos = np.unique(bernoulli, return_counts=True)
ax[0].bar(valores, conteos/len(bernoulli))
ax[0].set_title("Bernoulli (n=1)")
ax[0].set_xticks([0, 1])
ax[0].set_xlabel("0 / 1")
ax[0].set_ylabel("Frecuencia relativa")

# Binomial (0 a 10)
ax[1].hist(binomial, bins=np.arange(-0.5, 11.5, 1), density=True)
ax[1].set_title("Binomial (n=10)")
ax[1].set_xticks(range(0, 11))
ax[1].set_xlabel("Número de éxitos")

plt.tight_layout()
plt.show()
```

## Distribución de Poisson

* Supongamos λ = 4 llamadas por minuto.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro Poisson
lam = 4

# Simulación de 1000 minutos
np.random.seed(42)
muestras = np.random.poisson(lam=lam, size=1000)

print("Media estimada:", np.mean(muestras))
print("Varianza estimada:", np.var(muestras))


# Graficamos
valores, conteos = np.unique(muestras, return_counts=True)
frecuencias = conteos / len(muestras)

plt.bar(valores, frecuencias)
plt.xlabel("Número de llamadas por minuto")
plt.ylabel("Frecuencia relativa")
plt.title("Distribución Poisson (λ = 4)")
plt.show()
```

## Uniforme Continua

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
a = 2
b = 8

np.random.seed(42)

# Simulamos 10.000 valores
muestras = np.random.uniform(a, b, 10000) #1000000

print("Media estimada:", np.mean(muestras))
print("Media teórica:", (a + b) / 2)


# Histograma (densidad)
plt.hist(muestras, bins=30, density=True, edgecolor='black')
plt.axhline(1/(b-a), color='red', linestyle='--', label='Densidad teórica')
plt.title("Distribución Uniforme Continua U(2,8)")
plt.xlabel("Valores")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```

```python
prob_estimada = np.mean((muestras >= 3) & (muestras <= 5))
print("Probabilidad estimada:", prob_estimada)
print("Probabilidad teórica:", (5 - 3) / (8 - 2))
```

## Normal en Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
mu = 100
sigma = 10

np.random.seed(42)

# Generamos datos
muestras = np.random.normal(mu, sigma, 10000)

# Graficamos
plt.hist(muestras, bins=40, density=True, edgecolor='black')
plt.title("Distribución Normal (μ=100, σ=10)")
plt.xlabel("Valores")
plt.ylabel("Densidad")
plt.show()

print("Media estimada:", np.mean(muestras))
print("Desviación estándar estimada:", np.std(muestras))
```

```python
plt.hist(muestras, bins=40, density=True, alpha=0.6)
plt.axvline(mu, color='red', label='μ')
plt.axvline(mu - sigma, color='green', linestyle='--', label='μ - σ')
plt.axvline(mu + sigma, color='green', linestyle='--', label='μ + σ')
plt.legend()
plt.show()
```

## Distribución normal estándar

```python
from scipy.stats import norm

# 1) Probabilidad acumulada: P(Z <= 1)
p1 = norm.cdf(1)
print("P(Z <= 1) =", p1)

# 2) Probabilidad en un intervalo: P(-1 <= Z <= 1)
p2 = norm.cdf(1) - norm.cdf(-1)
print("P(-1 <= Z <= 1) =", p2)

# 3) Cola derecha: P(Z > 2)
p3 = 1 - norm.cdf(2)
print("P(Z > 2) =", p3)

# 4) Percentil: ¿qué z deja el 95% a la izquierda?
z95 = norm.ppf(0.95)
print("z del 95% =", z95)
```

```python
from scipy.stats import norm

mu = 100
sigma = 10

# P(X <= 120)
p = norm.cdf(120, loc=mu, scale=sigma)
print("P(X <= 120) =", p)

# Intervalo: P(90 <= X <= 110)
p_intervalo = norm.cdf(110, loc=mu, scale=sigma) - norm.cdf(90, loc=mu, scale=sigma)
print("P(90 <= X <= 110) =", p_intervalo)
```

## Distribución chi-cuadrado (χ²)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

x = np.linspace(0, 30, 400)

for df in [1, 2, 5, 10]:
    plt.plot(x, chi2.pdf(x, df=df), label=f"df={df}")

plt.title("Distribución Chi-cuadrado: cambia con los grados de libertad")
plt.xlabel("x")
plt.ylabel("densidad")
plt.legend()
plt.show()
```

## Distribución t de Student

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

x = np.linspace(-5, 5, 600)

dfs = [5, 20, 200]  # n=5 -> df=4 (pequeña), n=30 -> df=29 (grande)

plt.plot(x, norm.pdf(x), label="Normal estándar N(0,1)")

for df in dfs:
    plt.plot(x, t.pdf(x, df=df), label=f"t-Student (df={df})")

plt.title("t-Student: n pequeño vs n grande (comparación con Normal)")
plt.xlabel("x")
plt.ylabel("densidad")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
mu = 0
N = 20000  # número de repeticiones

def simular_t(n, mu=0, N=20000):
    # muestras desde una normal (población normal)
    X = np.random.normal(loc=mu, scale=1, size=(N, n))
    xbar = X.mean(axis=1)
    s = X.std(axis=1, ddof=1)  # s muestral
    t_stat = (xbar - mu) / (s / np.sqrt(n))
    return t_stat

t_pequena = simular_t(n=5, mu=mu, N=N)    # df=4
t_grande  = simular_t(n=30, mu=mu, N=N)   # df=29

plt.hist(t_pequena, bins=80, density=True, alpha=0.6, label="t simulada (n=5, df=4)")
plt.hist(t_grande,  bins=80, density=True, alpha=0.6, label="t simulada (n=30, df=29)")

plt.title("Distribución de la estadística t: n pequeño vs n grande")
plt.xlabel("t")
plt.ylabel("densidad")
plt.legend()
plt.show()
```

## Distribución de Fisher (F)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

x = np.linspace(0.001, 5, 600)  # evita 0 exacto

# Ejemplos de grados de libertad
pares_df = [(1, 10), (5, 10), (10, 10), (5, 30)]

for df1, df2 in pares_df:
    plt.plot(x, f.pdf(x, dfn=df1, dfd=df2), label=f"df1={df1}, df2={df2}")

plt.title("Distribución F de Fisher (densidad)")
plt.xlabel("x")
plt.ylabel("densidad")
plt.legend()
plt.show()
```

## Distribución exponencial

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

lam = 0.5  # tasa: 0.5 eventos por minuto (en promedio 1 evento cada 2 minutos)
N = 10000

# Simulación: tiempos de espera
muestras = np.random.exponential(scale=1/lam, size=N)

plt.hist(muestras, bins=50, density=True, edgecolor="black")
plt.title(f"Distribución Exponencial (λ={lam})")
plt.xlabel("Tiempo de espera")
plt.ylabel("Densidad")
plt.show()

print("Media estimada:", np.mean(muestras))
print("Media teórica:", 1/lam)
```
