**Deﬁnir eventos aleatorios y su representación con árboles de probabilidad**

Como trabajamos con el estudio sobre hábitos de salud en estudiantes universitarios, definimos el espacio muestral como:

						S="Estudiantes universitarios de la institución"

Definimos los siguientes eventos aleatorios a modo de ejemplo a fin de representar lo solicitado:

	A = “El estudiante realiza actividad física regular”

	B = “El estudiante duerme menos de 7 horas”


Estos eventos son variables aleatorias binarias (ocurre / no ocurre). Estos eventos se consideran eventos aleatorios porque:

- No sabemos de antemano si un estudiante seleccionado cumplirá la condición.					
- El resultado depende del azar asociado al proceso de muestreo.						
- Cada estudiante tiene cierta probabilidad de pertenecer al evento.			
		
Desde el punto de vista estadístico, cada evento puede modelarse como una variable aleatoria binaria (Bernoulli), que toma valores:			

1=ocurre el evento
0=no ocurre						

Además, En esta etapa no estamos suponiendo independencia entre los eventos. Ya que, en la realidad, realizar actividad física podría estar relacionado con dormir menos o más de 7 horas.

Por lo tanto, el árbol de probabilidad se construye permitiendo dependencia entre eventos.


<u>Definimos las probabilidades asociadas</u>

P(A)=Es la proporción esperada de estudiantes físicamente activos (realizan actividad física regular) en la población

P(B)=Probabilidad de que un estudiante seleccionado al azar duerma menos de 7 horas. Representa la frecuencia relativa de estudiantes que duerme menos de 7 horas.           

<u>Probabilidades complementarias</u>

P(A^c)=Es la probabilidad de que el estudiante no realice actividad física regular.
se define como: P(A^c)=1-P(A)

P(B^c)=Es la probabilidad de que el estudiante duerma más de 7 horas.
se define como: P(B^c)=1-P(B)

<u>Probabilidades conjuntas</u>

P(B∣A)=Probabilidad de que un estudiante duerma menos de 7 horas, dado que realiza actividad física regular
P(B∣A)=(P(A∩B))/(P(A))

P(B∣A^c)=Probabilidad de dormir menos de 7 horas dado que no realiza actividad física regular

<u>Probabilidad conjunta (intersección)</u>

P(A∩B)=Probabilidad de que un estudiante realice actividad física regular Y duerma menos de 7 horas
Se calcula mediante la regla del producto:
P(A∩B)=P(A)⋅P(B∣A) --> considerando dependencia
**En este punto podemos también calcularlo con el árbol de decisión**

<u>Probabilidad de unión</u>        

P(A∪B)=Probabilidad de que ocurra al menos uno de los eventos. esto es, realiza actividad física regular o duerme menos de 7 horas o ambas.

se calcula:             
P(A∪B)=P(A)+P(B)-P(A∩B)

**Árbol de decisión**

```python
from graphviz import Digraph

dot = Digraph()

# Nodo inicial
dot.node("I", "Estudiante\nseleccionado al azar")

# Primer nivel (Actividad física)
dot.node("A", "Actividad física regular (A)\nP(A) = p")
dot.node("Ac", "Actividad física no regular (Aᶜ)\nP(Aᶜ) = 1 - p")

dot.edge("I", "A")
dot.edge("I", "Ac")

# Segundo nivel (Horas de sueño)
dot.node("AB", "Duerme < 7h (B)\nP(B|A) = q")
dot.node("ABc", "Duerme ≥ 7h (Bᶜ)\nP(Bᶜ|A) = 1 - q")

dot.node("AcB", "Duerme < 7h (B)\nP(B|Aᶜ) = r")
dot.node("AcBc", "Duerme ≥ 7h (Bᶜ)\nP(Bᶜ|Aᶜ) = 1 - r")

dot.edge("A", "AB")
dot.edge("A", "ABc")

dot.edge("Ac", "AcB")
dot.edge("Ac", "AcBc")

dot
```

En este punto como aún no se crea el dataset simulado (en el punto 3), se definen las probabilidades como letras a fin de que sean representadas:               

P(A)=p          

P(B∣A)=q            

P(B∣A^c)=r

El árbol de decisión permite que graficamente podamos observar las probabilidades y la dependencia entre los enventos. Por ejemplo podemos decir, que la probabilidad de que un estudiante seleccionado al azar realice actividad física regular y duerma 7 horas o más es de p*(1-q), o que la probabilidad de que un alumno no realice actividad física regular y duerma menos de 7 horas es (1-p)*r.

Este punto nos permite un primer acercamiento para la realización del punto 4 que consiste en calcular probabilidades básicas de eventos (intersección, unión, complementarios) del dataset a simular.


En este estudio, el árbol facilita visualizar la secuencia de eventos, en este ejemplo, actividad física → horas de sueño de los estudiantes, a diferenciar probabilidades simples, condicionales y conjuntas y a analizar posibles relaciones o dependencias entre variables.
Desde la perspectiva de la Ciencia de Datos, esta herramienta permite modelar procesos estocásticos de forma clara, facilita la interpretación de probabilidades condicionales. Además, sirve como base conceptual para modelos más avanzados como bayesianos y árboles de decisión utilizados en Machine Learning. Asimismo, permite mejorar el razonamiento bajo incertidumbre.

**Simular el diseño muestral (con mínimo 100 registros)**

```python
# ============================================
# PROYECTO: Hábitos saludables universitarios
# Simulación reproducible para Inferencia Estadística
# ============================================

import numpy as np
import pandas as pd

# Generador moderno con semilla fija (reproducible)
rng = np.random.default_rng(42)

# Tamaño muestral
n = 150

# ----------------------------
# 1. Variables demográficas
# ----------------------------

edad = rng.integers(18, 26, n)

sexo = rng.choice(
    ["Femenino", "Masculino"],
    size=n,
    p=[0.55, 0.45]
)

# ----------------------------
# 2. Horas de sueño (Normal)
# ----------------------------

horas_sueno = rng.normal(loc=6.6, scale=1.1, size=n)
horas_sueno = np.clip(horas_sueno, 4, 9)
horas_sueno = np.round(horas_sueno, 1)

# ----------------------------
# 3. Nivel de estrés (1-10)
# ----------------------------

nivel_estres = rng.integers(1, 11, n)

# ----------------------------
# 4. Días de actividad física (Poisson)
# ----------------------------

dias_actividad = rng.poisson(lam=3, size=n)
dias_actividad = np.clip(dias_actividad, 0, 7)

# Variable derivada (Bernoulli)
actividad_regular = (dias_actividad >= 3).astype(int)

# ----------------------------
# 5. Calidad de alimentación (1-5)
# ----------------------------

calidad_alimentacion = rng.integers(1, 6, n)

# ----------------------------
# 6. Consumo frutas/verduras (Poisson)
# ----------------------------

consumo_frutas_verduras = rng.poisson(lam=2, size=n)

# ----------------------------
# 7. Crear DataFrame
# ----------------------------

df = pd.DataFrame({
    "Edad": edad,
    "Sexo": sexo,
    "Horas_sueno": horas_sueno,
    "Nivel_estres": nivel_estres,
    "Dias_actividad": dias_actividad,
    "Actividad_regular": actividad_regular,
    "Calidad_alimentacion": calidad_alimentacion,
    "Consumo_frutas_verduras": consumo_frutas_verduras
})

# Mostrar primeras filas
df.head()
```

Horas de sueño → simuladas con distribución normal (para trabajar TLC e intervalos).

Días de actividad → simulada con Poisson (conteo semanal).

Actividad regular → variable Bernoulli derivada (dias_actividad >= 3).

Variables categóricas → generadas con probabilidades controladas.

Semilla fija → garantiza reproducibilidad científica.

**Calcular probabilidades básicas de eventos (intersección, unión, complementarios)**

Primero calculamos las probabilidades simples para los 3 eventos principales que definimos:


	A = “El estudiante realiza actividad física regular”

	Su = “El estudiante duerme menos de 7 horas”

	E = “El estudiante presenta alto nivel de estrés”


```python
P_A = df["Actividad_regular"].mean()
print(f"la probabilidad del que el estudiante seleccionado realice actividad física regular es: {P_A:.4f}")

P_Su = (df["Horas_sueno"] < 7).mean()
print(f"la probabilidad del que el estudiante seleccionado duerma menos de 7 horas es: {P_Su:.4f}")

P_E= (df["Nivel_estres"] >= 7).mean()
print(f"la probabilidad del que el estudiante seleccionado tenga un nivel de estrés mayor o igual a 7 es: {P_E:.4f}")   
```

En este punto se observan las probabilidad individuales de los principales eventos definidos

<u>Probabilidades complementarias<u>

```python
P_Ac = 1 - P_A
print(f"la probabilidad del que el estudiante seleccionado no realice actividad física regular es: {P_Ac:.4f}")
P_Su_c = 1 - P_Su
print(f"la probabilidad del que el estudiante seleccionado duerma 7 horas o más es: {P_Su_c:.4f}")
P_Ec = 1 - P_E
print(f"la probabilidad del que el estudiante seleccionado tenga un nivel de estrés menor o igual a 6 es: {P_Ec:.4f}")
```

<u>Intersecciones</u>

Probabilidad de que el estudiante No realice actividad física regular y duerma menos de 7 horas:

esto nos da un primer acercamiento para ver la proporción de estudiantes que cumple con estas dos condiciones que se asocian en la hipótesis de estudio.

```python
P_A_inter_Su = (
    (df["Actividad_regular"] == 0) &
    (df["Horas_sueno"] < 7)
).mean()


print(f"la probabilidad del que el estudiante seleccionado NO realice actividad física regular y duerma menos de 7 horas es: {P_A_inter_Su:.4f}")
```

Probabilidad que el estudiante NO realice actividad física regular y tenga alto nivel de estrés (>=7):

```python
P_A_inter_E = (
    (df["Actividad_regular"] == 0) &
    (df["Nivel_estres"] >= 7)
).mean()
print(f"la probabilidad del que el estudiante seleccionado NO realice actividad física regular y tenga un nivel de estrés mayor o igual a 7 es: {P_A_inter_E}")
```

Probabilidad que el estudiante duerma menos de 7 horas y tenga alto nivel de estrés (>=7):

```python
P_Su_inter_E = (
    (df["Horas_sueno"] < 7) & (df["Nivel_estres"] >= 7)
    
).mean()


print(f"la probabilidad del que el estudiante seleccionado duerma menos de 7 horas y tenga un nivel de estrés mayor o igual a 7 es: {P_Su_inter_E:.4f}")
```

<u>Probabilidad condicional</u>

```python
P_Su_dado_A = (
    df[df["Actividad_regular"] == 1]["Horas_sueno"] < 7
).mean()
print(f"la probabilidad del que el estudiante seleccionado duerma menos de 7 horas dado que realiza actividad física regular es: {P_Su_dado_A:.4f}")

P_Su_c_dado_A = (
    df[df["Actividad_regular"] == 1]["Horas_sueno"] >= 7
).mean()
print(f"la probabilidad del que el estudiante seleccionado duerma 7 horas o más dado que realiza actividad física regular es: {P_Su_c_dado_A:.4f}")
P_Su_dado_Ac = (
    df[df["Actividad_regular"] == 0]["Horas_sueno"] < 7
).mean()
print(f"la probabilidad del que el estudiante seleccionado duerma menos de 7 horas dado que NO realiza actividad física regular es: {P_Su_dado_Ac:.4f}")

P_E_dado_A = (
    df[df["Actividad_regular"] == 1]["Nivel_estres"] >= 7
).mean()    
print(f"la probabilidad del que el estudiante seleccionado tenga un nivel de estrés mayor o igual a 7 dado que realiza actividad física regular es: {P_E_dado_A:.4f}")

P_E_dado_Ac = (
    df[df["Actividad_regular"] == 0]["Nivel_estres"] >= 7
).mean()
print(f"la probabilidad del que el estudiante seleccionado tenga un nivel de estrés mayor o igual a 7 dado que NO realiza actividad física regular es: {P_E_dado_Ac:.4f}")

P_E_dado_Su = (
    df[df["Horas_sueno"] < 7]["Nivel_estres"] >= 7
).mean()
print(f"la probabilidad del que el estudiante seleccionado tenga un nivel de estrés mayor o igual a 7 dado que duerme menos de 7 horas es: {P_E_dado_Su:.4f}")

P_E_dado_Su_c = (
    df[df["Horas_sueno"] >= 7]["Nivel_estres"] >= 7
).mean()
print(f"la probabilidad del que el estudiante seleccionado tenga un nivel de estrés mayor o igual a 7 dado que duerme 7 horas o más es: {P_E_dado_Su_c:.4f}")

pS_dado_E= (
    df[df["Nivel_estres"] >= 7]["Horas_sueno"] < 7
).mean()
print(f"la probabilidad del que el estudiante seleccionado duerma menos de 7 horas dado que tiene un nivel de estrés mayor o igual a 7 es: {pS_dado_E:.4f}")
```

<u>Probabilidad unión</u>

```python
P_union_A_Su = P_A + P_Su - P_A_inter_Su
print(f"la probabilidad de que un estudiante realice actividad física regular o duerma menos de 7 horas es: {P_union_A_Su}")

P_union_A_E = P_A + P_E - P_A_inter_E
print(f"la probabilidad de que un estudiante realice actividad física regular o tenga un nivel de estrés mayor o igual a 7 es: {P_union_A_E}")

P_union_Su_E = P_Su + P_E - P_Su_inter_E
print(f"la probabilidad de que un estudiante duerma menos de 7 horas o tenga un nivel de estrés mayor o igual a 7 es: {P_union_Su_E:.4f}")
```

**Árbol de probabilidad**

```python
from graphviz import Digraph

dot = Digraph()

# Nodo inicial
dot.node("I", "Inicio")

# Primer nivel (Actividad física)
dot.node("A", f"Actividad física regular (A)\nP(A)={P_A:.2f}")
dot.node("Ac", f"Actividad física no regular (Aᶜ)\nP(Aᶜ)={P_Ac:.2f}")

dot.edge("I", "A")
dot.edge("I", "Ac")

# Segundo nivel (Horas de sueño)
dot.node("AB", f"< 7 horas (Su)\nP(Su|A)={P_Su_dado_A:.2f}")
dot.node("ABc", f"≥ 7 horas (Suᶜ)\n{1-P_Su_dado_A:.2f}")

dot.node("AcB", f"< 7 horas (Su)\nP(Su|Aᶜ)={P_Su_dado_Ac:.2f}")
dot.node("AcBc", f"≥ 7 horas (Suᶜ)\n{1-P_Su_dado_Ac:.2f}")

dot.edge("A", "AB")
dot.edge("A", "ABc")

dot.edge("Ac", "AcB")
dot.edge("Ac", "AcBc")

dot
```

Ahora ya con el dataset se puede concluir, por ejemplo, que la probabilidad de que un estudiante seleccionado al azar realice actividad física regular y duerma menos de 7 horas es de (0.55x0.61), o que la probabilidad de que un alumno No realice actividad física regular y duerma 7 horas o más es de (0.45x0.31). 

Este árbol consituye un ejemplo ya que se puede realizar uno con el nivel de estrés para determinar las probabilidades conjuntas.

<b>Aplicar distribuciones adecuadas según el tipo de variable y contexto.</b>

**Identiﬁcar si las variables se ajustan a distribuciones discretas (binomial, Poisson) o continuas (normal) y Justiﬁcar la elección de las funciones de distribución.**

Analizamos las variables clave del estudio:

<u>Horas de sueño</u>	
				
Tipo: Cuantitativa continua                     
Modelo teórico: Distribución Normal

                                        X ∼ N (μ=6.6,σ=1.1)

**Justificación**
- Variable continua
- Se generó explícitamente con distribución normal
- Biológicamente el sueño suele aproximarse a distribución normal, se supone una variable continua y simétrica
- Permite inferencia sobre la media poblacional

La distribución normal es apropiada cuando la variable:
- Tiene simetría aproximada
- Presenta concentración alrededor de la media
- No está fuertemente sesgada


<u>Días de actividad física (a la semana)</u>             

Tipo: Cuantitativa discreta                 
Modelo teórico: Distribución Poisson            

                                        X ∼ Poisson(λ=3)

**Justificación**
- Cuenta eventos en intervalo fijo (7 días)       
- Variable discreta no negativa       
- Modela frecuencia de ocurrencias
- En este caso se cuentan eventos por unidad de tiempo            

La Poisson es adecuada cuando:
- Los eventos ocurren independientemente
- Se cuenta número de eventos en tiempo fijo


<u>Actividad regular (variable derivada)</u>            

Definida como:

                                 Actividad_regular= 1 si Dias_actividad ≥ 3
                                                    0 si Dias_actividad < 3

**1.-Generar distribuciones muestrales de media a partir del dataset simulado y 2.-Veriﬁcar empíricamente el Teorema del Límite Central**

**Tomar muchas muestras de tamaño n, calcular la media de cada muestra y ver cómo se distribuyen esas medias**          

Tomo muchas muestras de tamaño n = 50

```python
import numpy as np

n_muestra = 50
n_simulaciones = 5000

medias = []

for _ in range(n_simulaciones):
    muestra = df["Dias_actividad"].sample(n=n_muestra, replace=True)
    medias.append(muestra.mean())

medias = np.array(medias)

plt.figure()
plt.hist(medias, bins=40, density=True)
plt.title("Distribución muestral de la media (n=50)")
plt.xlabel("Media de días de actividad")
plt.ylabel("Densidad")
plt.show()

print("Media de las medias:", medias.mean())
print("Desviación estándar de las medias:", medias.std())
```

A partir de 5000 muestras de tamaño 50, se construyó la distribución muestral de la media de días de actividad física. Se observa que, a pesar de que la variable original es discreta y asimétrica, la distribución de las medias presenta una forma aproximadamente normal, lo cual **confirma empíricamente el Teorema del Límite Central**. Además, la media de las medias se acerca a coincidir con la media poblacional, evidenciando que la media muestral es un estimador insesgado.

**El Teorema del Límite Central dice:**


                                    X ∼ N (μ,σ/√n )

**Calculamos el error estándar teórico:**

```python
sigma = df["Dias_actividad"].std()
sigma_teorico = sigma / np.sqrt(n_muestra)

print("Error estándar teórico:", sigma_teorico)
print("Error estándar simulado:", medias.std())
```

El error estándar teórico fue calculado utilizando la desviación estándar poblacional del dataset y el tamaño muestral n=50. Se obtuvo un valor similar al observado en la simulación, confirmando que la dispersión de la distribución muestral se aproxima a σ/√n, tal como establece la teoría estadística.

```python
sigma = df["Dias_actividad"].std(ddof=0)

n = np.array([10, 25, 50, 100, 200])
se = sigma / np.sqrt(n)

plt.figure()
plt.plot(n, se)
plt.xlabel("Tamaño de muestra (n)")
plt.ylabel("Error estándar (σ/√n)")
plt.title("Disminución del error estándar al aumentar n")
plt.show()
```

<u>Para variable Horas_sueno</u>

Usamos t de Student porque no conocemos σ poblacional



Creo una función que permita calcular intervalos de confianza considerando la t-Student

```python
from scipy import stats

def intervalo_confianza(data, confianza):
    n = len(data)
    media = np.mean(data)
    s = np.std(data, ddof=1)
    error_estandar = s / np.sqrt(n)
    
    alpha = 1 - confianza
    t_critico = stats.t.ppf(1 - alpha/2, df=n-1)
    
    margen_error = t_critico * error_estandar
    
    return media, media - margen_error, media + margen_error, margen_error
```

Fijo los niveles para calcular los distintos Intervalos de Confianza

```python
niveles = [0.90, 0.95, 0.99]

for nivel in niveles:
    media, inf, sup, error = intervalo_confianza(df["Horas_sueno"], nivel)
    print(f"Horas sueño - IC {int(nivel*100)}%: ({inf:.4f}, {sup:.4f}) | Ancho: {(sup-inf):.4f}")
```

**Análisis:**

Con un 90% de confianza, la media poblacional de horas de sueño se encuentra entre 6.4783 y 6.7750 horas.

Con un 95% de confianza, la media poblacional de horas de sueño se encuentra entre 6.4496 y 6.8038 horas.

Con un 95% de confianza, la media poblacional de horas de sueño se encuentra entre 6.3928 y 6.8605 horas.

En cuanto al **ancho del intervalo:**       

90% → 0.2967

95% → 0.3542

99% → 0.4677

Eso demuestra que, a mayor confianza, mayor incertidumbre permitida en la estimación, esto quiere decir, que 
- Si busco estar más segura → acepto intervalo más amplio.              
- Si quiero más precisión → bajo nivel de confianza.                            

el cual es un 'Trade-off' clásico en inferencia.

**Interpretación**

Para la variable horas de sueño se construyeron intervalos de confianza al 90%, 95% y 99%. Se observa que la amplitud del intervalo aumenta conforme aumenta el nivel de confianza, pasando de 0.2967 a 0.4677 horas. Esto ocurre debido al incremento del valor crítico de la distribución t, lo que amplía el margen de error y, por tanto, la incertidumbre asociada a la estimación.

Los resultados indican que la media poblacional de horas de sueño se sitúa aproximadamente en torno a 6.6 horas, valor que se mantiene por debajo de 7 horas incluso al 99% de confianza. Esto sugiere que, en promedio, los estudiantes presentan un nivel de descanso inferior al recomendado.

Desde la perspectiva del estudio, este resultado es relevante porque refuerza la evidencia de que podrían existir factores individuales que estén influyendo negativamente en los hábitos saludables, principalmente en la variable analizada horas de sueño; en coherencia con la hipótesis general planteada. Además, el hecho de que todos los intervalos, incluso el más conservador (99%), se mantengan bajo el umbral de 7 horas aporta consistencia y solidez a la conclusión, disminuyendo la probabilidad de que el resultado se deba al azar.

En términos inferenciales, una estimación precisa y consistente como esta fortalece las bases para las pruebas de hipótesis posteriores (Lección 6), ya que reduce la incertidumbre sobre el parámetro poblacional y permite contrastar con mayor claridad las diferencias asociadas a variables como el nivel de estrés.

<font color="red">**HIPÓTESIS 1:**</font>
Estrés y sueño (media poblacional)          

•	H₀: No existe diferencia significativa en el promedio de horas de sueño según el nivel de estrés.                   
•	H₁: Existe diferencia significativa en el promedio de horas de sueño según el nivel de estrés.          

En este punto, se debe convertir el nivel de estrés en dos grupos comparables, por ejemplo:

Estrés bajo-moderado: 1–6

Estrés alto: 7–10

**Qué vamos a comparar**

Estamos comparando dos medias poblacionales:

μ₁ = media de horas de sueño en estudiantes con estrés bajo         

μ₂ = media de horas de sueño en estudiantes con estrés alto         

Lo que queremos saber es:                   
                        μ₁-μ₂ = 0?

Esto es ¿La diferencia entre ambas medias es estadísticamente distinta de cero?.                
            - Si la diferencia es pequeña → probablemente es azar.            
            - Si es grande → probablemente hay efecto real.

Se aplica un **test t para dos muestras independientes** porque: 

- Tenemos una variable cuantitativa continua: Horas_sueno → variable numérica continua.

- Tenemos dos grupos distintos: Estrés bajo y estrés alto.

- Son independientes: Un estudiante pertenece solo a un grupo.No es el mismo individuo medido dos veces.


 Aunque el tamaño muestral es mayor a 30, se utilizó la distribución t de Student debido a que la desviación estándar poblacional es desconocida y debe ser estimada a partir de la muestra. Para tamaños muestrales grandes, la distribución t converge a la normal, por lo que los resultados son muy similares, pero metodológicamente el uso de t es el procedimiento correcto.

● Se divide la muestra en dos subconjuntos.                 
● Se extrae las horas de sueño de cada grupo.               
● Se calcula el estadístico t.                  



**Calculo estadístico t y valor p**

```python
from scipy import stats

# Crear grupos
estres_bajo = df[df["Nivel_estres"] <= 6]["Horas_sueno"]
estres_alto = df[df["Nivel_estres"] > 6]["Horas_sueno"]

# Test t de dos muestras independientes
t_stat, p_valor = stats.ttest_ind(estres_bajo, estres_alto, equal_var=False)

print("Estadístico t:", f"{t_stat:.4f}")
print("Valor p:", f"{p_valor:.4f}")
```

Se aplicó un test t para muestras independientes con el objetivo de contrastar si existía una diferencia estadísticamente significativa en el promedio de horas de sueño entre estudiantes con bajo y alto nivel de estrés. Este procedimiento evalúa si la diferencia observada entre las medias muestrales puede atribuirse al azar o si refleja una diferencia real en las medias poblacionales. Dado que se desconoce la desviación estándar poblacional y se comparan dos grupos independientes, se utilizó la distribución t de Student.

**Interpretación:**         

El test t para muestras independientes arrojó un estadístico t = 0.0812 y un valor p = 0.9354. Dado que el p-valor es ampliamente superior al nivel de significancia α = 0.05, no se rechaza la hipótesis nula. En consecuencia, no se encontró evidencia estadística suficiente para afirmar que el nivel de estrés influya en el promedio de horas de sueño en la población estudiada.

Este resultado es consistente con la estructura del dataset simulado, en el cual ambas variables fueron generadas de manera independiente, sin un mecanismo causal explícito. Desde una perspectiva de análisis de datos, la ausencia de significancia refleja coherencia entre el modelo generativo y la inferencia estadística obtenida.

Estamos evaluando si la diferencia observada entre proporciones muestrales podría explicarse solo por variabilidad aleatoria.

Utilizo test de diferencia de proporciones (Z), atendido que:

● La variable respuesta es dicotómica (0/1).                
● Tenemos dos grupos independientes.                
● Se busca comparar probabilidades.

Cuando comparamos dos proporciones → usamos estadístico Z.

A diferencia del test t: Aquí la distribución muestral se aproxima a normal por el TLC aplicado a proporciones.

**La fórmula es:**



Donde el error estándar es:



**Además debe cumplirse que:**

El estadístico Z mide cuántos errores estándar separan las dos proporciones observadas.             

- Z grande → diferencia importante.                       
- Z cercano a 0 → proporciones similares.

```python
from statsmodels.stats.proportion import proportions_ztest

# Tabla de éxitos
tabla = df.groupby("Sexo")["Actividad_regular"].agg(["sum", "count"])

exitos = tabla["sum"].values
nobs = tabla["count"].values

z_stat, p_valor = proportions_ztest(exitos, nobs)

print("Estadístico Z:", z_stat)
print("Valor p:", p_valor)
```

Se aplicó un test de diferencia de proporciones para evaluar si la probabilidad de realizar actividad física regular difiere según el sexo. El estadístico Z compara la diferencia entre proporciones muestrales respecto al error estándar conjunto.

El test de diferencia de proporciones arrojó un estadístico Z = 1.6785 y un valor p = 0.0932. Dado que el p-valor es mayor que el nivel de significancia convencional α = 0.05, no se rechaza la hipótesis nula, por lo que no existe evidencia estadística suficiente para afirmar que la proporción de estudiantes que realiza actividad física regular difiere entre hombres y mujeres en la población estudiada. No obstante, el valor p obtenido no es elevado, sino relativamente cercano al umbral del 5%, lo que sugiere la presencia de una posible señal débil o tendencia. De hecho, si se considerara un nivel de significancia menos estricto (α = 0.10), la diferencia sería estadísticamente significativa, lo que evidencia que la decisión depende del criterio adoptado.

Desde una perspectiva de ciencia de datos, esto indica que podría existir un efecto pequeño que no alcanza significancia bajo el estándar del 5%, posiblemente debido al tamaño muestral o a la magnitud limitada del efecto. <font color="red">**HIPÓTESIS 3:**</font> Calidad de alimentación y nivel de actividad física

Aquí ya no se compara medias ni proporciones simples, sino que, se evalua la asociación entre dos variables categóricas.
- Calidad_alimentacion → ordinal (1 a 5)          
- Actividad_regular → dicotómica (0/1)                

En este punto, ambas variables se tratan como variables categóricas para este test.

Atendido que se busca saber si dos variables categóricas están asociadas, utilizo **Chi-cuadrado de independencia** porque:

● Observaciones independientes          
● Tamaño muestral suficientemente grande            
● Frecuencias esperadas mayores a 5 en al menos el 80% de las celdas            
● Además, no estamos comparando medias, no estamos modelando una variable continua; estamos comparando frecuencias categóricas.


En síntesis, se busca saber si la distribución de estudiantes activos y no activos es independiente del nivel de calidad de alimentación.                 
En términos probabilísticos si:                

                                    P(Actividad∣Alimentación)=P(Actividad)

• Si se cumple → hay independencia.                     
• Si no se cumple → hay asociación.                 

**Estadístico de prueba Chi-cuadrado**



Donde Oi  representa a cada frecuencia observada y ei representa a cada frecuencia esperada.

```python
from scipy.stats import chi2_contingency

tabla = pd.crosstab(df["Calidad_alimentacion"],
                    df["Actividad_regular"])

chi2, p_valor, dof, expected = chi2_contingency(tabla)

print("Chi-cuadrado:", f"{chi2:.4f}")
print("Grados de libertad:", dof)
print("Valor p:", f"{p_valor:.4f}")
```

El test de independencia χ² arrojó un estadístico de 4.05 con 4 grados de libertad y un valor p de 0.399 (Si realmente no existiera asociación entre alimentación y actividad física en la población, habría un 39.9% de probabilidad de observar diferencias como las encontradas (o mayores) solo por variabilidad aleatoria). Dado que el valor p es superior a α = 0.05, no se rechaza la hipótesis nula de independencia. En consecuencia, no se encontró evidencia estadística suficiente para afirmar que exista asociación entre la calidad de alimentación y el nivel de actividad física en la muestra analizada. Este resultado sugiere que, en el dataset utilizado, ambas variables se comportan de manera independiente.
