**● ¿Qué es el aprendizaje no supervisado?**

El aprendizaje de máquina no supervisado es una rama del machine learning que se utiliza para analizar datos que no poseen etiquetas o variables objetivo predefinidas. A diferencia del aprendizaje supervisado, donde el modelo aprende a partir de ejemplos con respuestas conocidas, en el aprendizaje no supervisado los algoritmos buscan descubrir patrones, estructuras o relaciones ocultas dentro de los datos sin una guía explícita.

El objetivo principal de este enfoque es comprender la estructura subyacente del conjunto de datos, identificando similitudes, agrupaciones o comportamientos recurrentes que no son evidentes a simple vista. Por esta razón, el aprendizaje no supervisado se utiliza con frecuencia en análisis exploratorio de datos, donde se busca obtener una primera comprensión de la información disponible antes de construir modelos predictivos.

Una de sus principales ventajas es que permite trabajar con grandes volúmenes de datos no etiquetados, los cuales son muy comunes en entornos reales como registros de clientes, interacciones en plataformas digitales, imágenes, texto o datos de sensores. Además, facilita la identificación de relaciones complejas o patrones ocultos, que pueden ser utilizados posteriormente para la toma de decisiones estratégicas o para mejorar modelos de análisis más avanzados.

No obstante, este tipo de aprendizaje también presenta ciertos desafíos. Debido a que no existe una respuesta correcta conocida, evaluar la calidad de los resultados puede ser más complejo, y la interpretación de los patrones encontrados requiere conocimiento del dominio del problema. Asimismo, la selección del algoritmo y de sus parámetros puede influir significativamente en los resultados obtenidos.

A pesar de estas limitaciones, el aprendizaje no supervisado constituye una herramienta fundamental dentro de la ciencia de datos moderna, especialmente cuando el objetivo es explorar, comprender y estructurar grandes conjuntos de datos antes de aplicar modelos más complejos.

**● Identiﬁcar las principales tareas que resuelve el aprendizaje no supervisado**

El aprendizaje no supervisado se utiliza para resolver diferentes tipos de problemas relacionados con la identificación de patrones en datos sin etiquetas. Entre las tareas más relevantes se encuentran la clusterización, la reducción de dimensionalidad, la detección de anomalías y el análisis de reglas de asociación.

<u>1. Clusterización (Clustering)</u>

La clusterización consiste en agrupar observaciones de manera que los elementos dentro de un mismo grupo sean más similares entre sí que con los de otros grupos. Este proceso permite identificar segmentos o estructuras naturales dentro de los datos sin necesidad de conocer previamente cuáles son esos grupos.
Esta técnica es ampliamente utilizada en diversos contextos, como por ejemplo:

- Segmentación de clientes según comportamiento de compra
- Agrupamiento de productos con características similares
- Análisis de comportamiento de usuarios en plataformas digitales
- Organización automática de imágenes o documentos

<u> 2. Reducción de dimensionalidad</u>         

La reducción de dimensionalidad busca disminuir la cantidad de variables o características presentes en un conjunto de datos, manteniendo la mayor cantidad posible de información relevante.
En muchos problemas reales, los datasets pueden contener decenas o cientos de variables, lo que dificulta su análisis y visualización. Mediante técnicas de reducción dimensional es posible simplificar la representación de los datos, facilitando tanto su interpretación como el procesamiento computacional.
Entre sus principales aplicaciones se encuentran:
- Visualización de datos complejos en dos o tres dimensiones
- Eliminación de redundancias entre variables
- Mejora del rendimiento (y eficiencia computacional) de algoritmos de machine learning
- Preparación de datos para otras tareas como clusterización

<u> 3. Detección de anomalías</u>

La detección de anomalías tiene como objetivo identificar observaciones que se comportan de forma significativamente diferente al resto de los datos. Estas observaciones atípicas pueden representar eventos inusuales o potencialmente problemáticos.
Este tipo de análisis es particularmente importante en contextos donde detectar comportamientos fuera de lo normal puede generar un gran valor, por ejemplo:                    
- Identificación de fraudes financieros         
- Detección de fallos en sistemas o sensores industriales           
- Identificación de transacciones sospechosas en comercio electrónico           
- Monitoreo de seguridad en redes informáticas      
- Identificación de comportamientos atípicos de usuarios                     


<u> 4. Análisis de reglas de asociación</u>             

El análisis de reglas de asociación busca descubrir relaciones frecuentes entre diferentes variables dentro de un conjunto de datos. En otras palabras, identifica qué eventos o elementos tienden a ocurrir juntos. A diferencia de la clusterización, que agrupa observaciones, estas técnicas se enfocan en descubrir patrones de co-ocurrencia entre elementos.
Este tipo de análisis es particularmente conocido en el ámbito del retail y el comercio electrónico, donde se utiliza para estudiar el comportamiento de compra de los clientes. Un ejemplo clásico corresponde al análisis de canasta de mercado, que identifica patrones como:

Los clientes que compran un determinado producto tienden a comprar también otro producto relacionado. Un ejemplo sería:                     

“Los clientes que compran pan también suelen comprar mantequilla”.

Este tipo de información permite a las empresas:                            
- Diseñar estrategias de ventas cruzadas (cross-selling)
- Optimizar la disposición de productos en tiendas
- Crear recomendaciones personalizadas en plataformas de e-commerce


Es importante destacar que estas tareas no son excluyentes y en muchos proyectos de ciencia de datos se utilizan de manera combinada. Por ejemplo, es común aplicar reducción de dimensionalidad antes de realizar una clusterización, con el objetivo de simplificar los datos y mejorar la calidad de los grupos obtenidos.

La selección de la técnica más adecuada dependerá del tipo de problema, los objetivos del análisis y las características del conjunto de datos. Una elección adecuada permite obtener resultados que no solo sean técnicamente correctos, sino también interpretables y útiles para la toma de decisiones.

Estas técnicas son muy utilizadas en análisis exploratorio de datos, ya que permiten identificar patrones y estructuras que serían difíciles de observar en espacios de alta dimensionalidad.

<u> 4. Técnicas de detección de anomalías</u>

Existen distintos enfoques para detectar anomalías en los datos. Entre los más utilizados se encuentran los siguientes:

**a) Métodos estadísticos**

Los métodos estadísticos se basan en el supuesto de que los datos siguen una determinada distribución probabilística. A partir de esta distribución, se identifican como anomalías aquellos valores que se encuentran muy alejados de la media o del comportamiento esperado. Estos métodos suelen utilizar medidas de dispersión y probabilidad para determinar qué observaciones pueden considerarse atípicas.

Entre las técnicas más utilizadas se encuentran:

- Z-score, que mide cuántas desviaciones estándar se encuentra un valor respecto de la media.           
- Test de Grubbs, utilizado para detectar outliers en muestras que siguen una distribución normal.

Estos métodos son especialmente útiles cuando se trabaja con datasets relativamente simples y distribuciones bien definidas.

**b) Métodos basados en distancia**

Los métodos basados en distancia identifican anomalías evaluando qué tan lejos se encuentra una observación respecto de sus vecinos más cercanos dentro del espacio de características. La idea central es que los puntos anómalos tienden a estar aislados o más alejados de otras observaciones, mientras que los datos normales suelen formar regiones más compactas.

Entre los algoritmos más conocidos se encuentran:

- k-Nearest Neighbors (k-NN) aplicado a detección de outliers       
- Local Outlier Factor (LOF), que compara la densidad local de un punto con la de sus vecinos

Estos métodos son útiles cuando los datos presentan estructuras complejas que no siguen distribuciones estadísticas simples.

**C) Métodos basados en densidad**

Los métodos basados en densidad detectan anomalías analizando la concentración de datos en diferentes regiones del espacio. Las observaciones que se encuentran en áreas donde la densidad de puntos es muy baja suelen considerarse anomalías. Estos enfoques son particularmente útiles cuando los datos presentan agrupaciones naturales o clústeres, ya que permiten identificar puntos que no pertenecen claramente a ningún grupo.

Un algoritmo representativo es:

- DBSCAN adaptado, que identifica automáticamente puntos de ruido (noise) cuando estos no cumplen con los criterios mínimos de densidad para formar parte de un clúster.

Este tipo de método es muy útil en datasets con estructuras irregulares o clusters de forma no lineal.

**d) Métodos basados en modelos**

Los métodos basados en modelos construyen primero un modelo que representa el comportamiento normal de los datos. Posteriormente, aquellas observaciones que presentan una desviación significativa respecto a ese comportamiento esperado son consideradas anomalías. Este enfoque es especialmente útil en problemas donde el patrón normal es relativamente estable y las anomalías representan desviaciones claras.

Entre los algoritmos más utilizados se encuentran:

- One-Class SVM, que aprende una frontera que encapsula las observaciones normales.             
- Autoencoders, redes neuronales que aprenden representaciones comprimidas de los datos y detectan anomalías cuando la reconstrucción presenta errores elevados.            
- Isolation Forest, que identifica anomalías aislando observaciones mediante particiones aleatorias del espacio de datos.

Estos métodos son particularmente efectivos en datasets complejos o de alta dimensionalidad, donde los enfoques tradicionales pueden ser menos eficientes.

**Explicar ventajas y desventajas de la técnica**

Entre sus principales <u>ventajas</u> destacan:

- Permite descubrir patrones ocultos en los datos (Mediante la agrupación de datos similares, se pueden obtener valiosos conocimientos ocultos que son útiles en diversos campos)

- Facilita la segmentación de poblaciones heterogéneas

- Permite no solo entender mejor los datos, sino también tomar decisiones basadas en comportamientos, tendencias o
características compartidas entre subconjuntos.

- Contribuye al análisis exploratorio de grandes datasets

- Puede utilizarse como etapa previa a otros modelos de machine learning

Sin embargo, también presenta ciertas limitaciones o <u>desventajas</u> que deben considerarse durante su aplicación:

- Necesidad de normalizar los datos para evitar que variables con escalas mayores dominen el análisis.

- Sensibilidad al método de distancia elegido, que puede afectar signiﬁcativamente los resultados.


- Algunos algoritmos son sensibles a outliers o a la escala de las variables.

- La elección de parámetros y número de clústeres puede influir significativamente en los resultados.

- Validar los resultados sin etiquetas puede ser complejo y requiere métodos adicionales como el índice de Silhouette

- La interpretación de los clústeres puede requerir conocimiento del dominio del problema.

Por esta razón, es habitual complementar la clusterización con métricas de evaluación y visualización de datos para validar la calidad de los grupos obtenidos.

**Implementar PCA y t-SNE con scikit-learn**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

Aplico estandarización

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

<font size=6>**PCA**</font>

Primero se realiza un análisis de la varianza explicada y acumulada

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

var_exp = pca.explained_variance_ratio_
var_acum = np.cumsum(var_exp)

# Mostrar resultados
resultados = pd.DataFrame({
    "Componente": [f"PC{i}" for i in range(1, len(var_exp) + 1)],
    "Varianza_Explicada": var_exp,
    "Varianza_Acumulada": var_acum
})

print(resultados)
```

El bajo porcentaje de varianza explicada por las primeras componentes principales sugiere que la estructura del dataset es altamente multidimensional, por lo que una proyección en 2D no logra capturar adecuadamente la complejidad de los datos.

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

**Método del codo K-Means**

Se evaluaron valores de K en un rango de 2 a 15, ya que permite capturar posibles estructuras en los datos sin generar una segmentación excesivamente fragmentada y que dificulte la interpretación.

```python
inertia = []

K_range = range(1, 16)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, marker='o')
plt.title("Método del Codo")
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.show()
```

**Coeficiente de Silhouette**

```python
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    print(f"K={k}, Silhouette Score={score:.3f}")
```

Aunque el método del codo sugiere una reducción progresiva hasta valores cercanos a K=15, este número resulta poco práctico desde el punto de vista interpretativo. Además, el coeficiente de silueta no presenta mejoras significativas si comparamos los k, desde el k=9 al k=15. Por esta razón, se opta por un número de clusters más reducido, que permite un mejor equilibrio entre calidad del agrupamiento y utilidad analítica.

Si bien el coeficiente de silueta continúa aumentando para valores mayores de K (Se hicieron pruebas con k mayores), este incremento es marginal y no justifica la complejidad adicional. Por ello, se considera más ópttima la elección de un valor intermedio que equilibra la calidad del agrupamiento y la interpretabilidad de los resultados.

**Coeficiente de Silhouette gráficamente**

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1) Usar tus datos reales
X = X_scaled

# 2) Probar distintos K
k_values = range(2, 11)
sil_scores = []

for k in k_values:
    kmeans = KMeans(
        n_clusters=k,
        init="k-means++",
        n_init=10,
        random_state=42
    )
    
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)

# 3) Graficar
plt.figure(figsize=(8, 5))
plt.plot(list(k_values), sil_scores, marker="o")
plt.title("Silhouette Score vs Número de Clusters")
plt.xlabel("Número de clusters (K)")
plt.ylabel("Silhouette promedio")
plt.xticks(list(k_values))
plt.grid(True, alpha=0.3)
plt.show()

# 4) Mejor K
best_k = list(k_values)[sil_scores.index(max(sil_scores))]
print(f"Mejor K según Silhouette: {best_k} (score={max(sil_scores):.3f})")
```

Aunque el coeficiente de silueta continúa aumentando a medida que crece el número de clusters (se hicieron más pruebas con k mayores), los valores más altos se alcanzan con un número elevado de K, lo que dificulta la interpretabilidad del modelo. Por esta razón y las mencionadas con anterioridad, se selecciona un valor intermedio K=12, que representa un equilibrio entre calidad de agrupamiento y utilidad práctica, evitando el sobreajuste.

**DBSCAN**

"Tunning" epsilon

```python
eps_values = np.linspace(0.5, 2.5, 10)  # puedes ajustar este rango
min_samples_values = [5, 10]
```

```python
results = []

for eps in eps_values:
    for min_samples in min_samples_values:
        
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(X)
        
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = np.sum(labels == -1)
        
        sil = None
        mask = labels != -1
        
        if len(set(labels[mask])) >= 2:
            sil = silhouette_score(X[mask], labels[mask])
        
        results.append([eps, min_samples, n_clusters, n_noise, sil])
```

```python
df_results = pd.DataFrame(
    results,
    columns=["eps", "min_samples", "n_clusters", "noise", "silhouette"]
)

df_results.sort_values(by="silhouette", ascending=False).head(20)
```

Aunque DBSCAN alcanza valores altos de coeficiente de silueta para ciertos parámetros, estos resultados se obtienen a costa de generar un gran número de clusters y una alta proporción de puntos clasificados como ruido. Esto indica que el dataset no presenta una estructura densa claramente definida, lo que limita la efectividad de este algoritmo para el problema analizado
